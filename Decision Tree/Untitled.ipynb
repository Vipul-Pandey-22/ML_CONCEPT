```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('label_disese', axis=1), df['label_disese'], test_size=0.2, random_state=42)
```

```python
X_train = X_train[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4']]
X_test = X_test[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4']]
```

# Fill NaN

```python
print(X_train['Symptom_4'].isna().sum())
print(X_test['Symptom_4'].isnull().sum())
```

```python
# Find mode 
print(X_train['Symptom_4'].mode()[0])
print(X_test['Symptom_4'].mode()[0])
```

```python
# Fill NaN
X_train['Symptom_4'].fillna(X_train['Symptom_4'].mode()[0], inplace=True)
X_test['Symptom_4'].fillna(X_test['Symptom_4'].mode()[0], inplace=True)
```

```python
X_train
```

# Apply OHE

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Assuming your DataFrame is named 'data' and symptom columns are ['symptom1', 'symptom2', 'symptom3', 'symptom4']
ohe = OneHotEncoder(drop='first', sparse_output=False)  # drop='first' to avoid multicollinearity

# Apply OneHotEncoder only on the symptom columns
column_transformer = ColumnTransformer([('ohe', ohe, ['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4'])], remainder='passthrough')

# Transform your data
transformed_train_data = column_transformer.fit_transform(X_train)
transformed_test_data = column_transformer.transform(X_test)
```

```python
transformed_train_data.shape
```

```python
transformed_test_data.shape
```

# Apply DT

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(transformed_train_data, y_train)
```

```python
y_pred = dt.predict(transformed_test_data)
```

```python
print(dt.score(transformed_test_data, y_pred))
```

```python
# Apply Hyperparameter Tuning
dt1 = DecisionTreeClassifier()
```

```python
param = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(dt1,param_grid=param,cv=5, scoring='neg_mean_squared_error')
```
