```python
df.head()
```

```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('Unknown')
mapping = {'Unknown': 0, 'Sleep Apnea': 1, 'Insomnia': 2}
df['Sleep'] = df['Sleep Disorder'].map(mapping)
df
```

```python
df = df.drop(columns=['Person ID', 'Sleep Disorder'], axis=1)
df
```

```python
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/',expand=True)

# Convert to numeric values
df['Systolic'] = pd.to_numeric(df['Systolic'])
df['Diastolic'] = pd.to_numeric(df['Diastolic'])

df = df.drop(columns=['Blood Pressure'], axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Sleep']), df['Sleep'], test_size=0.3)
```
